```julia
function newton_raphson(f,f´,xi,e,n)
    
    raiz=Inf;
    it=0;
    
    for i in 1:n
        xi=xi-f(xi)/f´(xi);
        it=it+1
        if (abs(f(xi))<=e)
            raiz=xi
            break
        end 
    end
    print("Raiz de la función=$raiz\n") 
    print("Numero de iteraciones realizadas=$it\n")
    print("Error absoluto de la imagen = $(f(raiz))\n")
    return xi
end
    
```

```julia
function Grafico(raiz,x,f)
    plot(x,f)
    plot!([(x[1]),(x[end])],[0,0],c="black",legend=false)
    scatter!([raiz],[0])
end
```

```julia
f(x)=sin(x)
f´(x)=cos(x)

@show raiz=newton_raphson(f,f´,4,1.0e-10,30)
using Plots
x= LinRange(1,10,200)
Grafico(raiz,x,f)
```

```julia
f(x) = x*x - x +2 
f´(x)=2*x-1

@show raiz=newton_raphson(f,f´,0,1.0e-10,30)
using Plots
x= LinRange(-1.5,10,200)
Grafico(raiz,x,f)
```

## Método de la Secante


```julia
x2, x1 = 1, 2           # initial guess of 2
## (1,2)
f(x) = x^2 - 2          # some function
fp(x1,x2) = (f(x1) - f(x2))/(x1 - x2)
x2, x1 = x2 - f(x2)/fp(x1, x2), x2 # update step
## (1.3333333333333333,1)
```

```julia
function m_secante(f,a,b,e,n)
    
    raiz=Inf;
    it=c=0;
    
    for i in 1:n
        c=b-f(b)*(a-b)/(f(a)-f(b));
        it=it+1
        if (abs(f(c))<=e)
            raiz=c
            break
        end 
        a=b;b=c;
    end
    print("Raiz de la función=$raiz\n") 
    print("Numero de iteraciones realizadas=$it\n")
    print("Error absoluto de la imagen = $(f(raiz))\n")
    return c
end
```

```julia
f(x) = sin(x)*log(x^2) 

raiz=m_secante(f,10,9,1.0e-9,1e2)
using Plots
x= LinRange(1,30,200)
Grafico(raiz,x,f)
```

```julia
f(x) = x*x - x +2  

raiz=m_secante(f,10,9,1.0e-9,1e2)
using Plots
x= LinRange(-1.5,30,200)
Grafico(raiz,x,f)
```
